```python
"""
activity 더미 엑셀 파일을 mysql에 넣기위한 프로그램
"""

import configparser
import pymysql
import xlrd

# 엑셀 읽기
wb = xlrd.open_workbook(filename = 'activity_db.xlsx')

cities = ["파리", "뉴욕", "도쿄", "타이페이", "다낭", "세부"]
activities = []
for city in cities:
    ws = wb.sheet_by_name(city)
    activity = []
    for r in range(ws.nrows):
        if r == 0:
            continue
        col = []
        for c in range(ws.ncols):
            if c == 1 or c == 6:
                col.append(int(ws.cell(r, c).value))
            else:
                col.append(ws.cell(r, c).value)
        activity.append(col)
    activities.append(activity)

# config 변수 정리
config = configparser.ConfigParser()
config.read('config.ini')

db_host = config['DATABASE']['db_host']
db_user = config['DATABASE']['db_user']
db_password = config['DATABASE']['db_password']
db_name = config['DATABASE']['db_name']

# 스키마와 연결
conn = pymysql.connect(host=db_host, 
                       user=db_user, 
                       password=db_password, 
                       db=db_name,
                       charset='utf8')
# 커서 생성
curs = conn.cursor()

# SQL문들
sql_insert_activity = "INSERT INTO Activity (name, cost, content, url_mrt, url_kl, img, CityId) VALUES(%s, %s, %s, %s, %s, %s, %s)"
sql_select_activity = "select * from Activity"
sql_update_activity = "UPDATE Activity SET cost = %s, urlHotel = %s WHERE cityId = %s AND category = %s"

    
# Activity 테이블의 내용을 가져오기 + 출력
curs.execute(sql_select_activity) 
activity_data = curs.fetchall() 

if len(activity_data) == 0: # activity table이 비어있을 경우, insert
    for city_activities in activities:
        curs.executemany(sql_insert_activity, city_activities)
else: # 이미 들어있으면 skip 
    pass

conn.commit()
```
